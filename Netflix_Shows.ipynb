```python
import matplotlib.pyplot as plt
import numpy as np
```

```python




shows = ["Stranger Things", "Squid Game", "Money Heist", "Wednesday", "Bridgerton",
         "The Witcher", "Lucifer", "Ozark", "You", "The Crown"]
viewers = [120, 111, 105, 98, 90, 85, 80, 75, 80, 68]
ratings = [8.7, 8.0, 8.3, 8.1, 7.7, 8.2, 8.1, 8.4, 7.8, 8.6]


x = np.arange(len(shows))
width = 0.4

plt.bar(x - width/2, viewers, width, label='Viewers (M)', color='crimson')
plt.bar(x + width/2, ratings, width, label='IMDb Rating', color='gold')


plt.xticks(x, shows, rotation=45, ha='right')
plt.ylabel("Value")
plt.title("Top 10 Netflix Shows: Viewers vs IMDb Ratings")
plt.legend()
plt.tight_layout()
plt.show()
```
