```python
import matplotlib.pyplot as plt
```

```python
houses = ["Gryffindor", "Slytherin", "Hufflepuff", "Ravenclaw"]
points = [550, 480, 420, 460]

# Pie chart colors and highlight
colors = ['red', 'green', 'yellow', 'blue']


plt.figure(figsize=(8,8))
plt.pie(points, labels=houses, autopct='%1.1f%%', startangle=140, colors=colors,)
plt.title("Hogwarts House Cup Points Distribution")
plt.show()
```
