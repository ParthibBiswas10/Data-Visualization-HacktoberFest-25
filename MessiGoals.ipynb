```python
import matplotlib.pyplot as plt
```

```python
teams = ["Brazil", "Germany", "France", "Bolivia", "Spain", "uruguay", "ecuador"]
goals = [5, 1, 3, 11, 2, 6, 7]


plt.figure(figsize=(11,6))
plt.stem(teams, goals, basefmt=" ", linefmt='skyblue', markerfmt='o', bottom=0)
plt.ylabel("Goals Scored")
plt.title("Lionel Messi's Goals Against International Teams")
plt.grid(axis='y', linestyle='--', alpha=0.9)
plt.show()
```
