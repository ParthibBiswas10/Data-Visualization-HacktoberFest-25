```python
import matplotlib.pyplot as plt
```

```python
genres = ["Fantasy", "Sci-Fi", "Romance", "Mystery", "Non-Fiction", "Thriller"]
readers = [120, 80, 100, 70, 90, 60]


plt.bar(genres, readers, color='skyblue')
plt.ylabel("Number of Readers")
plt.title("Number of Readers per Genre")
plt.show()
```
